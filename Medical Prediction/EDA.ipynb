```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cardio = pd.read_csv("cardio_train.csv",sep=';')
cardio.head()
```

```python
cardio.info()
```

```python
cardio.describe()
```

```python
labels = ['presence', 'absence']
explode = [0.02,0.02]
plt.pie(cardio['cardio'].value_counts(),autopct='%1.2f%%',labels=labels,colors=['darkslategray','powderblue'],
        explode=explode,startangle=90);
plt.title("Presence or absence of cardiovascular disease in patients",fontsize=10);
```

```python
labels = ['normal', 'above normal', 'well above normal']
explode = [0.02, 0.02, 0.02]

plt.pie(cardio['cholesterol'].value_counts(), labels=labels, explode=explode,autopct='%1.1f%%', colors=['darkslategrey','cadetblue','powderblue']);
plt.title("Cholesterol levels in patients",fontsize=10, loc='center');
```

```python
sns.histplot(data=[i/365 for i in cardio['age']],bins=100, color='powderblue');
plt.title("Age distribution of patients", fontsize=10, loc='center');
plt.xlabel("Age(years)");
```

```python
explode = [0.02, 0.02]
labels = ['Non-smokers', 'smokers']

plt.pie(cardio['smoke'].value_counts(),labels=labels, explode=explode,autopct='%1.1f%%',colors=['darkslategray','powderblue']);
plt.title("Proportion of smokers in patients",fontsize=10,loc='center');
```

```python
sns.histplot(cardio['weight'],bins=100,color='powderblue');
plt.title("Weight distribution of patients",fontsize=10,loc='center');
```

```python
sns.displot(cardio['height'],bins=100,color='cadetblue');
plt.title("Height distribution of patients",fontsize=10,loc='center');
plt.xlabel("Height(cm)");
```

```python
labels = ['women','men']
explode = [0.02,0.02]
plt.pie(cardio[cardio['cardio']==1]['gender'].value_counts(),autopct='%1.2f%%',labels=labels,colors=['darkslategray','powderblue'],
        explode=explode,startangle=90);
plt.title("Gender distribution of patients",fontsize=10);
```
